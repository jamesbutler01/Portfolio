# Dynamic information encoding by neural networks

In this project, we're going to look at real neuron data from the anterior cingulate cortex (ACC), that is well known for encoding the value of objects that we encounter. We shall use correlations and PCA to visualise the exact temporal dynamics of how the neural network encodes information.

Data are taken from a task where the subjects viewed a cue worth one of five, equally spaced, values. ~800 ms later a second cue was shown, which either added or competed with the first cue. Hence they needed to add or subtract the first value from the second value to figure out the total amount of reward. 

#### Data import

Our x data consists of the value of the image on each trial and our y data is the pre-processed normalised firing rate of ACC neurons for 0-1500 ms after the first cue was shown. 

```python
import numpy as np
import scipy
import plot  # Custom plot function for this script
import matplotlib.pyplot as plt
import maths
import sklearn.decomposition
import seaborn as sb
from mpl_toolkits.mplot3d import Axes3D

x = np.load('x.npy')[0] # The value of the cue [numFeatures x numSamples]
y = np.load('y.npy')  # The firing rates of our neurons over time [numFeatures x numSamples x numTimepoints]

# Data parameters
time = range(0, 1800, 10)  # Data is in 10 ms steps
numNeurons, numSamples, numTimepoints = y.shape
numValues = 5  # There were 5 different values in total
```

#### Data exploration

Lets first examine how individual neurons behave in response to the different value cues. We'll bin trials by the value of the cue that was shown and calculate the average and SEM of the firing rates for each bin over time.

```python
averages,sems = np.empty((numNeurons,numValues,numTimepoints)),np.empty((numNeurons,numValues,numTimepoints))

for i, (yy, xx) in enumerate(zip(y, x)):  # For each neuron
        
    # Calculate mean and SEM for response to each value
    averages[i] = [np.mean(yy[xx==val], axis=0) for val in np.unique(xx)]
    sems[i] = [np.std(yy[xx==val], axis=0) / np.sqrt(len(yy[xx==val])) for val in np.unique(xx)]

plot.plot1(averages, sems);
```

Cool, we can see our neurons are really sensitive to value. But they are also very diverse in how they encode value. Some increase their firing rates for high values (bottom row) and some for low values (top row). The time course of each response also differs, both in latency and duration.  

Let's now use linear regression to model each neuron's response to value at each time point. I've already covered this in another [project](https://github.com/jamesbutler01/Portfolio/blob/main/Simultaneous%20value%20coding/NeuronValueCoding.ipynb), so we'll use the custom functions from there.

```python
coeffs = np.empty((numNeurons, numTimepoints))

for i, (yy, xx) in enumerate(zip(y, x)):  # For each neuron
        
    # Fit model and log coefficients
    coeffs[i] = maths.reg(xx, yy)[1]

plot.plot2(coeffs)
```

Great, we can see a real heterogeneous mix of coefficients across our neural population, and this is just one timepoint!

#### Pearson r and correlation matrices

Lets now examine how the coefficients across the network change over time. To do this we'll compute a correlation matrix to examine how similar the value response is between the different time points. Because time is autocorrelated which will confound our analysis, we'll split our data into two random halves and fit our model on each half seperately. The correlation of coefficients between these two halves is then free of any autocorrelation confounds.

```python
coeffs = np.empty((2, numNeurons, numTimepoints))
for i, (yy, xx) in enumerate(zip(y, x)):  # For each neuron
        
    # Random 50:50 split
    randInds = np.random.permutation(numSamples)
    firstsplit,secondsplit = randInds[::2],randInds[1::2]

    # Fit model and log coefficients
    coeffs[0, i] = maths.reg(xx[firstsplit], yy[firstsplit])[1]
    coeffs[1, i] = maths.reg(xx[secondsplit], yy[secondsplit])[1]

# Correlate our coefficients over time
corrMatrix = np.corrcoef(coeffs[0].T, coeffs[1].T)

# Just want the top right quarter of the corrMatrix
corrMatrix = corrMatrix[numTimepoints:,:numTimepoints]

plot.plot3(corrMatrix)
```

Cool, we can see the coefficients were very correlated between our two splits of the data (as the r values are very high overall). This demonstrates that the neural network produces the same stereotyped response to value across trials. 

Lets now use Scipy's pearson function to get the corresponding p-values for our r values and only display the significant portions of our correlation matrix.

```python

def calcSigCorrMatrix(coeffs):
    # Time by time point correlation to get each p-value
    rs, ps = np.empty([numTimepoints]*2),np.empty([numTimepoints]*2)
    for i in range(numTimepoints):
        for j in range(numTimepoints):
            rs[i,j],ps[i,j] = scipy.stats.pearsonr(coeffs[0,:,i],coeffs[1,:,j])

    # NaN any points where p>0.05
    rs[ps>0.05] = np.nan
    
    return rs, ps

sigCorrMatrix, pvalues = calcSigCorrMatrix(coeffs)

plot.plot3(sigCorrMatrix, 1)
```

Now we can see that the correlations are focused around the diagonal of the matrix whereas the off diagonal components are mostly uncorrelated. Therefore, the neural network's encoding is dynamic - the way in which it represents information constantly changes over time such that no two time points are similar.

#### The neural network can rotate its encoding

An additional part of the task is that a second cue was shown ~800 ms after the first cue. This cue either added or subtracted to the value of the first cue and thus affected how the stored value of the first cue should be handled. So far we've been averaging over this effect by examining all trials. Let's now split trials by this condition and examine the correlation over time between the two cue types.

```python
# Load whether the second cue added or subtracted from the first cue
trialtype = np.load('x.npy')[1]

coeffs = np.empty((2, numNeurons, numTimepoints))
for i, (yy, xx, t) in enumerate(zip(y, x, trialtype)):

    # Fit model seperately for each trial type and store coefficients
    coeffs[0, i] = maths.reg(xx[t==0], yy[t==0])[1]
    coeffs[1, i] = maths.reg(xx[t==1], yy[t==1])[1]

    
corrMatrix, pvalues = calcSigCorrMatrix(coeffs)

plot.plot3(corrMatrix,2)
```

Strikingly, after the second cue was shown (black dashed line) the network's value coefficients are no longer correlated between the two trial types. Let's look at the time point where this happens in a bit more detail.

```python
tp = 95  # 950 ms after cue 1 (~150 ms after cue 2)
r, p = scipy.stats.pearsonr(coeffs[0,:,tp],coeffs[1,:,tp])

# Plot
sb.regplot(coeffs[0,:,tp], coeffs[1,:,tp], label=f'r={np.round(r, 4)}\np={np.round(p, 4)}')
plt.xlabel('Second cue additive'), plt.ylabel('Second cue subtractive'); plt.legend()
plt.title(f'Cue 1\'s coefficients at {tp*10} ms');
```

We can see that shortly after the second cue was shown, the coefficients for cue 1 are no longer correlated with one another. Instead they are now  orthogonal to one another. This means that the neural network had the ability to rotate its stored information into two orthogonal subspaces depending on what it needed to do with the stored information. 

#### Principal component analysis

Principal component analysis (PCA) is a powerful tool that allows us to transform high dimensional data into lower dimensional subspaces. We'll now use PCA to visualise the network rotating its stored information into these two orthogonal subspaces. 

The input to the PCA will be the neural firing rates in response to each of the five values in our two conditions (so 10 features in total). Lets get those values now.

```python
avgs = np.empty((10, numNeurons, numTimepoints))

# Loop through each cell's data
for i, (yy, xx, t) in enumerate(zip(y, x, trialtype)):
        
    # Calculate average response to each value for first trial type
    avgs[:5, i] = [np.mean(yy[(xx==val) & (t==0)], axis=0) for val in np.unique(xx)]

    # Same again for second trial type
    avgs[5:, i] = [np.mean(yy[(xx==val) & (t==1)], axis=0) for val in np.unique(xx)]
```

Now PCA only works in 2 dimensions, but we currently have 3 dimensions ([values x neurons x time]). So we'll average the time dimension over our point of interest to create a matrix with dimensions [values x neurons]. Then we can perform PCA on this matrix.

```python
# Average over time dimension
valuesByNeurons = np.mean(avgs[:, :, 50:80],axis=-1)

# Do PCA
pca = sklearn.decomposition.PCA()
model = pca.fit(valuesByNeurons)

#Plot
plt.plot( model.explained_variance_ratio_*100); plt.title('Variance explained by each PC');
plt.xlabel('Principal component');plt.ylabel('% Variance explained');
```

As is typical with biological data, the majority of the variance in our neuron firing rates is explained by the first couple of  principal components (PCs). 

We can now project our original timecourse data back onto a low dimensional subspace defined by the first two PCs. This will enable us to visualise how the network's encoding of the first cue's value rotates depending on the type of second cue shown.

```python
dotProducts = np.empty((10, 2, numTimepoints))

for i in range(10):  # For each value in each condition
    
    # Take the dot product of the PCA weights with the neuron firing rates
    dotProducts[i, :] = [np.dot(model.components_[ii], avgs[i]) for ii in range(2)]

plot.plot4(dotProducts)
```

Cool, we can see how initially the trajectories for each value in the two conditions are near to one another. Then, when the second cue is shown, the trajectories rapidly diverge between the two conditions. 

Let's look at two time points more closely to better visualise the effect. 

```python
# Take timepoint before and after second cue occurred
timepoint1 = 30; timepoint2 = 80
timepoints = dotProducts[..., [timepoint1, timepoint2]]

plot.plot5(timepoints)
```

Both before and after the second cue is shown, we can see that PC1 encodes the value of the cue (value follows the x-axis of the plots above). This is true for both conditions (solid and dashed lines).

After the second cue was shown, interestingly, PC2 encodes opposite values depending on the trial type (reading along the y-axis in the bottom plot gives you opposing values for each condition). This is a perfect coding scheme for handling adding and subtracting, as the first cue's value has opposite effects between the two operations! A downstream region can therefore use the first PC if it simply wants to know the value of the first cue, or the second PC if it wants to add or subtract the first cue's value. 

This is a novel demonstration of how the ACC neural network is able to dynamically alter its stored value representations depending on what we want to use the information for.
